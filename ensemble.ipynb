```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)
```

```python
def preprocess_data(data):
    # Fill missing values
    data['Age'].fillna(data['Age'].median(), inplace=True)
    data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
    data['Fare'].fillna(data['Fare'].median(), inplace=True)

    # Encode categorical variables
    label_encoder = LabelEncoder()
    data['Sex'] = label_encoder.fit_transform(data['Sex'])
    data['Embarked'] = label_encoder.fit_transform(data['Embarked'])

    # Select relevant features
    features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
    X = data[features]
    y = data['Survived']
    return X, y

X, y = preprocess_data(data)
```

```python
# Initialize models
dt_clf = DecisionTreeClassifier(random_state=42)
lr_clf = LogisticRegression(max_iter=200, random_state=42)
knn_clf = KNeighborsClassifier(n_neighbors=5)
```

```python
# Voting Classifier (Integrating the models)
voting_clf = VotingClassifier(
    estimators=[('dt', dt_clf), ('lr', lr_clf), ('knn', knn_clf)],
    voting='hard'  # 'hard' for majority voting, 'soft' for averaging probabilities
)
```

```python
# Train the Voting Classifier
voting_clf.fit(X, y)
```

```python
# Make predictions
y_pred = voting_clf.predict(X)
```

```python
# Evaluate the model
accuracy = accuracy_score(y, y_pred)
print("Accuracy of Voting Classifier by Ensembler Technique:", accuracy)
```
